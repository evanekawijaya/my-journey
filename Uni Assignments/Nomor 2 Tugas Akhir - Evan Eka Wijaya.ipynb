```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data_corona={'Tanggal':list(range(1,32)),"Jumlah Kasus":[433,292,349,395,484,367,338,336,533,387,233,4484,689,568,490,529,489,496,486,693,973,634,949,526,479,415,686,687,678,557,700]}
data=pd.DataFrame(data_corona)
data['Jumlah Kumulatif']=data["Jumlah Kasus"].cumsum()

data.plot.line(x='Tanggal', y='Jumlah Kumulatif', figsize=(20, 8), style='ro-',label='Jumlah total kasus sampai pada hari tersebut')
plt.title("Pertumbuhan Kasus COVID-19 di Indonesia pada Bulan Mei 2020", fontsize=20)
plt.ylabel("Jumlah kasus")
plt.xlabel("Tanggal")
plt.xticks(np.arange(1,32,1))
for x,y in zip(data['Tanggal'],data['Jumlah Kumulatif']):
    plt.annotate(y,(x,y),textcoords="offset points", xytext=(0,10),ha='center')
plt.legend()

data.plot(x='Tanggal',y='Jumlah Kasus',kind='bar',figsize=(20,8),color='green')
plt.title('Jumlah Kasus COVID-19 Baru di Indonesia per Hari pada Bulan Mei 2020', fontsize=20)
plt.ylabel('Jumlah kasus baru')
plt.xlabel('Tanggal')
plt.yticks(np.arange(0,5000,200))
plt.show()
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)

# Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')

# Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')

# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')

# Note that this is the same as
plt.subplots(2, 2, sharex=True, sharey=True)

# Create figure number 10 with a single subplot
# and clears it if it already exists.
fig, ax = plt.subplots(num=10, clear=True)
```
